1 List all competitions along with their category name

```python
from tabulate import tabulate
```

```python
mycursor = connection.cursor()

    
query = ("""
        SELECT competitions.competition_name AS competition_name, Categories.category_name 
        FROM competitions
        INNER JOIN Categories ON competitions.category_id = Categories.category_id;
    """)
mycursor.execute(query)
results = mycursor.fetchall()

    
print(tabulate(results, headers=["Competition Name", "Category Name"], tablefmt="psql"))
```

2 Count the number of competitions in each category

```python
mycursor = connection.cursor()

    
query = ("""
        SELECT Categories.category_name, COUNT(Competitions.category_id) AS competition_count
        FROM Competitions
        INNER JOIN Categories ON Competitions.category_id = Categories.category_id
        GROUP BY Categories.category_name;
    """)
mycursor.execute(query)
results = mycursor.fetchall()

    
print(tabulate(results, headers=["Competition Name", "Category Name"], tablefmt="psql"))
```

3 Find all competitions of type 'doubles'

```python
mycursor = connection.cursor()
query = ("""
        SELECT Competitions.competition_name AS competition_name, Categories.category_name
        FROM Competitions
        INNER JOIN Categories ON Competitions.category_id = Categories.category_id
        WHERE Categories.category_name = 'Juniors';
    """)
mycursor.execute(query)
results = mycursor.fetchall()
print(tabulate(results, headers=["Competition Name", "Category Name"], tablefmt="psql"))
```

6 Analyze the distribution of competition types by category

```python
mycursor = connection.cursor() 
query = ("""
        SELECT 
            Categories.category_name,
            Competitions.type,
            COUNT(*) AS competition_count
        FROM Competitions
        INNER JOIN Categories ON Competitions.category_id = Categories.category_id
        GROUP BY Categories.category_name, Competitions.type
        ORDER BY Categories.category_name, competition_count DESC;
    """)

mycursor.execute(query)
results = mycursor.fetchall()
print(tabulate(results, headers=["Category Name", "Competition Type", "Number of Competitions"], tablefmt="psql"))
```

List all venues along with their associated complex name
Count the number of venues in each complex
Get details of venues in a specific country (e.g., Chile)
Identify all venues and their timezones
Find complexes that have more than one venue
List venues grouped by country
Find all venues for a specific complex (e.g., Nacional)


```python
##List all venues along with their associated complex name
mycursor = connection.cursor()
query = ("""
        SELECT venues.venue_name,
            complexes.complex_name
        FROM venues
        INNER JOIN complexes ON venues.complex_id = complexes.complex_id;
    """)

mycursor.execute(query)
results = mycursor.fetchall()

    # Display results in tabular format
print(tabulate(results, headers=["Venue Name", "Complex Name"], tablefmt="psql"))
```

```python
#Count the number of venues in each complex
mycursor = connection.cursor()

    
query = ("""
        SELECT 
            complexes.complex_name,
            COUNT(venues.venue_id) AS venue_count
        FROM venues
        INNER JOIN complexes ON venues.complex_id = complexes.complex_id
        GROUP BY complex_name;
    """)
mycursor.execute(query)
results = mycursor.fetchall()

    
print(tabulate(results, headers=["Complex Name", "Number of Venues"], tablefmt="psql"))
```

```python
#Get details of venues in a specific country (e.g., Chile)
mycursor = connection.cursor()

query = ("SELECT country_name,venue_id,venue_name,city_name,country_code,timezone from venues WHERE country_name='japan' ")

mycursor.execute(query)
results = mycursor.fetchall()


print(tabulate(results, headers=["Country_Name","Venue_id","Venue_name","City_name","Country_code","Timezone"], tablefmt="psql"))
```

```python
##Identify all venues and their timezones
mycursor = connection.cursor()

query = ("SELECT venue_name, timezone FROM venues")

mycursor.execute(query)
results = mycursor.fetchall()


print(tabulate(results, headers=["Venue_Name","Timezone"], tablefmt="psql"))
```

```python
###Find complexes that have more than one venue
mycursor = connection.cursor()

query = ("""SELECT complex_id FROM (SELECT COUNT(venue_id) as venue_id_count, complex_id FROM venues GROUP BY complex_id) 
                 AS childquery
                  WHERE venue_id_count>1 """)

mycursor.execute(query)
results = mycursor.fetchall()


print(tabulate(results, headers=["Complex_id"], tablefmt="psql"))
```

```python
#List venues grouped by country
mycursor = connection.cursor()

query = ("SELECT country_name, GROUP_CONCAT(venue_name) AS venues FROM venues GROUP BY country_name ;")
mycursor.execute(query)
results = mycursor.fetchall()


print(tabulate(results, headers=["country_name","venue_names"], tablefmt="psql"))
```

```python
###Find all venues for a specific complex (e.g., Nacional)
mycursor = connection.cursor()

query = ("""SELECT venue_name,complex_name 
        FROM (SELECT venues.venue_name,complexes.complex_name
        FROM venues
        inner join complexes ON venues.complex_id=complexes.complex_id)
          AS joined_table WHERE complex_name=LOWER('Khalifa International Tennis & Squash Complex')""")
mycursor.execute(query)
results = mycursor.fetchall()


print(tabulate(results, headers=["venue_name","complex_name"], tablefmt="psql"))
```

Get all competitors with their rank and points.
Find competitors ranked in the top 5
List competitors with no rank movement (stable rank)
Get the total points of competitors from a specific country (e.g., Croatia)
Count the number of competitors per country
Find competitors with the highest points in the current week


```python
##Get all competitors with their rank and points.


mycursor = connection.cursor()

query = ("""SELECT competitors.competito_name,competitor_rankings.ranks,competitor_rankings.points
                FROM competitors inner join competitor_rankings ON 
                 competitors.competitor_id=competitor_rankings.competitor_id""")
mycursor.execute(query)
results = mycursor.fetchall()


print(tabulate(results, headers=["Competitor_Nma","Ranks","Points"], tablefmt="psql"))
```

```python
#Find competitors ranked in the top 5
mycursor = connection.cursor()

query = ("""
    SELECT competitors.competito_name, competitor_rankings.ranks
    FROM competitors
    INNER JOIN competitor_rankings ON competitors.competitor_id = competitor_rankings.competitor_id
    ORDER BY competitor_rankings.ranks ASC
    LIMIT 5;
""")

mycursor.execute(query)
results = mycursor.fetchall()


print(tabulate(results, headers=["Competitor_Name","Ranks"], tablefmt="psql"))
```

```python
##List competitors with no rank movement (stable rank)
mycursor = connection.cursor()

query = ("""SELECT competitors.competito_name, competitor_rankings.ranks
FROM competitors
INNER JOIN competitor_rankings 
    ON competitors.competitor_id = competitor_rankings.competitor_id
WHERE competitor_rankings.ranks = competitor_rankings.ranks
ORDER BY ranks""")

mycursor.execute(query)
results = mycursor.fetchall()


print(tabulate (results, headers=["Competitor_Name","Ranks"] ,tablefmt="psql"))
```

```python
#Get the total points of competitors from a specific country 
mycursor = connection.cursor()
query=("""SELECT  country, sum(points) AS sum_of_points FROM (SELECT competitors.country,competitor_rankings.points
                FROM competitors inner join competitor_rankings ON 
                 competitors.competitor_id=competitor_rankings.competitor_id) AS NEWTABLE WHERE country = 'USA' """)
mycursor.execute(query)
results = mycursor.fetchall()


print(tabulate (results, headers=["county","points"] ,tablefmt="psql"))
```

```python
#Count the number of competitors per country
mycursor = connection.cursor()
query = """SELECT competitors.country, COUNT(competito_name) AS Total_competitor 
           FROM competitors 
           GROUP BY country 
           ORDER BY Total_competitor DESC"""
mycursor.execute(query)
results = mycursor.fetchall()

print(tabulate(results, headers=["Country", "Total Competitors"], tablefmt="psql"))
```

```python
##Find competitors with the highest points in the current week
mycursor = connection.cursor()
query = ("SELECT max(points) AS APEX_POINT from competitor_rankings")
mycursor.execute(query)
results = mycursor.fetchall()

print(tabulate(results, headers=["Country", "Total Competitors"], tablefmt="psql"))
```
